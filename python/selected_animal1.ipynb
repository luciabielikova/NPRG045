```python
import math

def euclidean_distance(point1, point2):
    return math.sqrt((point1[0] - point2[0]) ** 2 + (point1[1] - point2[1]) ** 2)


def get_geometry(name):
    a = animals[animals["name"] == name]["geometry"];
    for coordinate in a:
        for point in coordinate.coords:
            return point

odkial = get_geometry("veverka")
pokial = get_geometry("sloni")

def get_all_points(): 
    all_points =[]
    for coordinate in highways['geometry']:
        for point in coordinate.coords:
            all_points.append(point)
    return all_points



def closest_point(given_point, points):
    min_distance = float('inf')
    closest = None
    for point in points:
        distance = euclidean_distance(given_point, point)
        if distance < min_distance:
            min_distance = distance
            closest = point
    return closest
```
